```python
import ppscore as pps
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
ASD = pd.read_csv('C:/Users/sanju/Capstone2Project/Autism_cleaned_data.csv', index_col=0)
```

```python
ASD.head()
```

```python
sns.catplot(x="Sex", y="Qchat-10-Score", hue="Class/ASD Traits ", data=ASD, kind="box", height=4, aspect=.5, 
            palette={'Yes': "r", 'No': "b"});
sns.catplot(x="A9", y="Qchat-10-Score", hue="Class/ASD Traits ", data=ASD, kind="box", height=4, aspect=.5, 
            palette={'Yes': "r", 'No': "b"});
```

```python
pps.score(ASD, "A9", "Class/ASD Traits ")
```

```python
matrix_df = pps.matrix(ASD)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
plt.figure(figsize = (10,10))
sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=0.5, annot=True)
```
